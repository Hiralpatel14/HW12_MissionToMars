# Mission to Mars

# Step 1 Scraping  NASA Mars News

```python
# import dependencies
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import pymongo
import time
import os
```

```python
# URL of page to be scraped
nasa_url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
response = requests.get(nasa_url)

# Create BeautifulSoup object; parse with 'lxml'
soup1 = BeautifulSoup(response.text, 'lxml')
type(soup1)
```

```python
print(soup1.body.prettify())
```

```python
# extract the latest news Title 
news_titles = soup1.find_all('div', class_='content_title')

# print title
print(news_titles)
```

```python
#loop to get all titles
for t in news_titles:
    title = t.find('a')
    title_names = title.text
    print(title_names)
```

```python
# get 1st news Titles
news_title1 = soup1.find_all('div', class_='content_title')[0].find('a').text.strip()
print(news_title1)
```

```python
# find the paragraph text
paragraphs = soup1.find_all('div', class_='rollover_description')
print(paragraphs)
```

```python
#loop through paragraph texts
for p in paragraphs:
    p_text = p.find('div')
    para_text = p_text.text
    print(para_text)
```

```python
# extract the latest 1st news paragraph text
news_p = soup1.find_all('div', class_ = 'rollover_description_inner')[0].text.strip()
print(news_p)
```

# JPL Mars Space Images - Featured Image

```python
jpl_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'

# Retrieve page with the requests module
response = requests.get(jpl_url)

# Create BeautifulSoup object; parse with 'lxml'
soup2 = BeautifulSoup(response.text, 'lxml')
type(soup2)
```

```python
# find image link with BeautifulSoup
images = soup2.find_all('div', class_='carousel_items')
images
```

```python
# loop through Images
for img in images:
    image = img.find('article')
    background_image = image.get('style')
    print(background_image)
```

```python
# extract url
re_background_image = re.search("'(.+)'",background_image)
print(re_background_image)
```

```python
search_background_image= re_background_image.group(1)
search_background_image
```

```python
featured_image_url = f'https://www.jpl.nas/gov{search_background_image}'
featured_image_url
```

```python
#extra to look
part_address = soup2.find_all('a',class_='fancybox')[0].get('data-fancybox-href').strip()
```

```python
#extra
Featured_img_url = "https://www.jpl.nas/gov"+part_address
print(Featured_img_url)
```

```python
# jpl another image to find
images1 = soup2.find_all('div', class_='grid_layout')
images1
```

```python
# loop through Images
for img in images1:
    image = img.find('article')
    background_image1 = image.get('slide')
    print(background_image1)
```
